```python
#importing the data 
import pandas as pd
df_firearm=pd.read_csv("firearm.csv")
len(df_firearm)
```

```python
#Exploring the dataset
df_firearm.head(10)
df_firearm.columns
df_firearm.isnull().sum()
#As it can be seen , there are some Nan Values in the dataset that need to be removed.
```

Before removing the Nan values, it would be better to remove the columns that are not needed for the analysis first, and then check the count of Nan values again. 

Columns like 'ISO code', 'Source', 'Rank by rate of ownership', '% of homicides by firearm', 'Homicide by firearm rate per 100,000 pop', 'Average firearms per 100 people' are not necessary for the analysis.


```python
#Removing the above mentioned columns
df_firearm.drop(['ISO code','Source','Rank by rate of ownership','% of homicides by firearm',\
                 'Homicide by firearm rate per 100,000 pop','Average firearms per 100 people'],axis=1,inplace=True)
df_firearm.head(5)
```

```python
#checking the number of Nan values 
df_firearm.isnull().sum()
```

```python
#Removing the Nan values
df_firearm.dropna(inplace=True)
df_firearm.head(10)
```

```python
df_firearm.isnull().sum()
```

To develop a decpetive version of the visualization , I took the following 2 steps:
1. Keep the top 14 countries with the maximum number of firearm homicides (There are 4 more countries above the USA.)
2. Remove the 4 countries on top of USA to show that USA has the highest number of firarm homicides

```python
##getting the top 14 countries
df_firearm1=df_firearm.sort_values(['Number of homicides by firearm'],ascending=[False]).head(14)
df_firearm1
```

```python
##removing the countries above USA
df_firearm_deception = df_firearm1.drop(df_firearm1.index[0:4]) 
df_firearm_deception
```

Now we can save this dataset to carry out deceptive visualization in Tableau
