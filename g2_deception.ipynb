#### (WORLD vs USA) Number of civilian owned guns compared to the Population :

```python
#Importing the original dataset for firearms and homicides
import pandas as pd
df_firearm_2=pd.read_csv("firearm_2.csv")
len(df_firearm_2)

#Renaming "Country/Territory" column to "Country"
df_firearm_2.rename(columns={'Country/Territory':'Country'},inplace=True)
df_firearm_2.head(5)
```

```python
#importing population data
import pandas as pd
df_population=pd.read_csv("population.csv")
len(df_population)
df_population.head(10)

```

```python
#Removing the first 3 columns, and taking 1st row as data and not header
df_population.drop(['Unnamed: 0','Unnamed: 1'],axis=1,inplace=True)
```

```python
df_population.head(5)
```

```python
#Merging the 2 dataframes on "Country" column
df_merged=df_population.merge(df_firearm_2,on="Country")
df_merged.head(10)
```

```python
#Dropping unnecessary columns
df_merged.drop(['ISO code','Source','% of homicides by firearm','Homicide by firearm rate per 100,000 pop',\
                'Rank by rate of ownership','Average firearms per 100 people','Number of homicides by firearm'],\
               axis=1,inplace=True)
```

```python
df_merged.head(10)
```

```python
#checking for null values 
df_merged.isnull().sum()

#removing the null values 
df_merged.dropna(inplace=True)
```

```python
len(df_merged)
```

Considering that the dataset is already in descending order in terms of the 'Population', for the Deceptive Visualization, I would like to see the top 10 most populted countries with their average total all civilian firearms.

```python
df_merged_pop=df_merged.head(10)
df_merged_pop
```

```python
#Now the population based dataset is ready to be visualized. Exporting it to a csv file ..

df_merged_pop.to_csv("df_merged_pop_deception.csv")
```

I would like to explore the countries in terms of their 'Average total all civilian firearms.'

```python
df_merged_fire=df_merged_pop.sort_values(['Average total all civilian firearms'],ascending=[False]).head(10)
df_merged_fire
```

```python
#Exporting the file to csv 
df_merged_fire.to_csv("df_merged_fire_deception.csv")
```
