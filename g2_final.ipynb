```python
##try to find % difference of population and civilian owned guns of each country
df_merged2['Population ']=df_merged2['Population '].str.replace(',','').astype(float)
df_merged2['Average total all civilian firearms'] = df_merged2['Average total all civilian firearms'].astype(float)
df_merged2['Percent Difference']=(df_merged2['Average total all civilian firearms']/df_merged2['Population '])*100
df_merged2

df_merged3=df_merged2.sort_values(['Percent Difference'],ascending=[False]).head(10)
df_merged3
```

```python
df_merged3.to_csv('percent_diff.csv')
```

Taking inspiration from the article itself which compares US with the rest of the world, I took US as a separate element and combined all the data of rest of the countries of the world. I combined both - the population data as well as the civilian owned guns data. 

```python
df_merged4=df_merged2.copy()
df_merged4=df_merged4.rename(columns={'Population ':'Population',\
                                      'Average total all civilian firearms':'Civilian Firearms(Average)'})
us_pop=df_merged4.loc[2,'Population']
us_arms=df_merged4.loc[2,'Civilian Firearms(Average)']


pop_diff=df_merged4['Population'].sum()-df_merged4.loc[2,'Population']
arms_diff=df_merged4['Civilian Firearms(Average)'].sum()-df_merged4.loc[2,'Civilian Firearms(Average)']


World_and_usa=pd.DataFrame(columns=['Country','Population','Civilian Firearms(Average)'])
rows = [['USA',us_pop,us_arms],['World',pop_diff,arms_diff]]
for row in rows:
    World_and_usa.loc[len(World_and_usa)] = row
print(World_and_usa)
World_and_usa.to_csv('world_usa_deception.csv')



```
